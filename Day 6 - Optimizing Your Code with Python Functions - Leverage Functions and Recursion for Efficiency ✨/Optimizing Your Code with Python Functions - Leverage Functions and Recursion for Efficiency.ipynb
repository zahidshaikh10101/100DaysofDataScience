# Functions in Python

Python Functions is a block of statements that return the specific task. The idea is to put some commonly or repeatedly done tasks together and make a function so that instead of writing the same code again and again for different inputs, we can do the function calls to reuse code contained in it over and over again.



```python
#Basic Hellow world function
def hello():
    print("Hello User!")
hello()
```

```python
#create a function that add 100 to user input number and display the result
def add_100():
    a = int(input("Enter a number: "))
    print(a + 100)
    
add_100()
```

## Functions with Arguments/Parameter

```python
# create a function to sum up three number given by user as parameter
def add_3_numbers(x,y,z):####arguments/parameters
    a=x+y+z
    print(a)

add_3_numbers(50,100,150)
```

```python
# create a functions that convert any word from lower_case to Upper_case
def upper_case(a):
    a = a.upper()
    print(a)
    
x = input("Enter any word: ")
upper_case(x)
```

```python
# create a functions that convert any word from Upper_case to lower_case
def lower_case(a):
    a = a.lower()
    print(a)
    
x = input("Enter any word: ")
lower_case(x)
```

```python
## create a function which will check wheather the number is multiple of 11
def m_o_11(x):
    if x%11==0:
        print(x,'is multiple of 11')
    else:
        print(x,'is not a multiple of 11')
        
m_o_11(77)  
```

```python
def palindrome(s):
    s=s.lower()
    c=s[::-1]
    if c==s:
        print(s,'is palindrome')
    else:
        print(s,'is not a palinmdrome')
        
s = "abcdcba"
palindrome(s)
```

```python
def factorial(x):
    sums = 1
    for i in range(x,0,-1):
        sums = sums * i
    return sums

n = int(input("Enter a number: "))
factorial(n)
```
