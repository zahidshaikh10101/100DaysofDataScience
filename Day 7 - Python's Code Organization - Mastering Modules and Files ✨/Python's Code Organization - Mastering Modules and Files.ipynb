# Modules in Python

# Math Module

```python
import math
```

```python
# Round a number upward to its nearest integer
print(math.ceil(1.4))
print(math.ceil(5.3))
print(math.ceil(-5.3))
```

```python
# Total number of possible combinations
n = 7
k = 5

print(math.comb(n, k))
```

```python
# Return factorial of a number
print(math.factorial(9))
print(math.factorial(6))
print(math.factorial(12))
```

```python
#Round numbers down to the nearest integer
print(math.floor(1.4))
print(math.floor(5.3))
print(math.floor(-5.3))
```

```python
#Returns the sum of all items in any iterable (tuples, arrays, lists, etc.)
print(math.fsum([1, 2, 3, 4, 5]))
print(math.fsum([100, 400, 340, 500]))
print(math.fsum([1.7, 0.3, 1.5, 4.5]))
```

```python
#Return the value of x raised to the power of y
print(math.pow(9, 3))
print(math.pow(7, 4))
print(math.pow(2, 8))
```

```python
pie = math.pi #Return the value of pie
print(pie)
```

# Calendar:
We can even print a calendar for a particular month. This can be done using the calendar module.

```python
import calendar
print(calendar.month(2023, 11))
```

```python
months = []
for i in range(1,13):
    months.append(i)

for i in months:
    print(calendar.month(2023,i))
```

# File Handling in Python
- Python too supports file handling and allows users to handle files i.e., to read and write files, along with many other file handling options, to operate on files.
- Python has several functions for creating, reading, updating, and deleting files.
- Python also supports different modes for opening files, including:
    - "r": Read (default mode). Opens the file for reading.
    - "w": Write. Opens the file for writing, truncating the file if it already exists.
    - "a": Append. Opens the file for writing, but appends new content to the end of the file.
    - "b": Binary mode. For working with binary files, like images or videos.
    - "x": Exclusive creation. Opens the file for writing, but only if the file doesn't already exist.

### Working in Read mode

```python
file = open("myfile.txt", "r") 
print (file.read())
```
