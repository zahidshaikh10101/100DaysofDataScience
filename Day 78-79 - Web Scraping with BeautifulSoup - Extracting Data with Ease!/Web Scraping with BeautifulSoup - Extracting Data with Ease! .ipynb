```python
from bs4 import BeautifulSoup
import requests
from urllib.parse import urljoin
import pandas as pd
import numpy as np
```

```python
base_url = "http://quotes.toscrape.com/"
response = requests.get(base_url)
print("Status Code: ", response.status_code)
print('-' * 152)
print("Metadata: ", response.headers)
```

```python
soup = BeautifulSoup(response.text, "html.parser")
print(soup.prettify())
```

```python
tag_links = []

# Extract links for tags
links = soup.find_all("a")
for i in links:
    href = i.get('href')
    if href and href.startswith("/tag/"):
        full_url = urljoin(base_url, href)  # Convert relative URL to absolute
        tag_links.append(full_url)

# Extract tag names and links for tag DataFrame
tag_data = []
for link in tag_links:
    tag_name = link.split("/tag/")[1].split("/")[0].replace("-", " ")  # Extract tag name
    tag_data.append({"category": tag_name, "link": link})

tag_df = pd.DataFrame(tag_data)

# Display the DataFrames
print("Tag DataFrame:")
tag_df
```

```python
author_links = []
# Extract links for authors
for i in links:
    href = i.get('href')
    if href and href.startswith("/author/"):
        full_url = urljoin(base_url, href)  # Convert relative URL to absolute
        author_links.append(full_url)
        
# Extract author names and links for author DataFrame
author_data = []
for link in author_links:
    author_name = link.split("/author/")[1].replace("-", " ")  # Extract and format author name
    author_data.append({"author": author_name.title(), "link": link})

author_df = pd.DataFrame(author_data)
print("\nAuthor DataFrame:")
print(author_df)
```

```python
import pandas as pd
from bs4 import BeautifulSoup
from urllib.parse import urljoin

# Sample URL (update `base_url` and soup accordingly)
base_url = "http://quotes.toscrape.com/"
response = requests.get(base_url)
soup = BeautifulSoup(response.text, 'html.parser')

# Extract quotes
quotes = soup.find_all("span", class_="text")

# Extract authors
authors = soup.find_all("small", class_="author")

# Extract links for authors
links = soup.find_all("a")
author_links = []
for i in links:
    href = i.get('href')
    if href and href.startswith("/author/"):
        full_url = urljoin(base_url, href)
        author_links.append(full_url)

# Extract tags
tags_data = []
quote_tags = soup.find_all("div", class_="tags")
for tag_block in quote_tags:
    tags = [tag.text for tag in tag_block.find_all("a", class_="tag")]
    tags_data.append(", ".join(tags))  # Join tags as a comma-separated string

# Combine all extracted data into a DataFrame
data = []
for idx, quote in enumerate(quotes):
    data.append({
        "Quotes": quote.text.strip(),
        "Author": authors[idx].text.strip() if idx < len(authors) else None,
        "Author_Url": author_links[idx] if idx < len(author_links) else None,
        "Tags": tags_data[idx] if idx < len(tags_data) else None
    })

df = pd.DataFrame(data)

# Display the resulting DataFrame
df
```
