### 1. Import numpy as np and see the version

```python
import numpy as np
np.__version__
```

### 2. How to create a 1D array?

```python
x = np.arange(10)
x
```

### 3. How to create a boolean array?

```python
np.full((4,3), True, dtype=bool)
```

```python
np.zeros((9), dtype=bool).reshape(3,3)
```

### 4. How to extract items that satisfy a given condition from 1D array?

```python
x = np.arange(30)
x[x%2==1]
```

### 5. How to replace items that satisfy a condition with another value in numpy array?

```python
x = np.arange(30)
x[x%2==1] = -1
x
```

### 6. How to get the common items between two python numpy arrays?

```python
a = np.array([1,2,3,2,3,4,3,4,5,6,7])
b = np.array([7,2,10,2,7,4,9,4,9,8])
np.intersect1d(a,b)
```

### 7. How to get all the items between two python numpy arrays?

```python
np.union1d(a,b)
```

### 8.How to extract all numbers between a given range from a numpy array?

```python
a = np.array([7,2,10,2,12,7,4,14,9,4,9,8])
a[(a>6) & (a<=10)]
```

### 9. How to reverse the rows of a 2D array?

```python
arr = np.arange(12).reshape(4,3)
print("Original Array: ")
print(arr)
print("\nModified Array: ")
print(arr[::-1,])
```

### 10. How to swap two columns in a 2d numpy array?

```python
arr = np.arange(20).reshape(4,5)

print('Original array')
print(arr)
print("\nModified array")
print(arr[:, [4,1,0,3,2]])
```
