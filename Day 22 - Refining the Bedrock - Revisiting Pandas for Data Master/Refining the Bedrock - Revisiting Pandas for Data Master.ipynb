```python
df1 = df.copy()
df1 = df1.rename(columns={"Emp_No":"Employee Number","Emp_Name":"Name","Emp_Age":"Age",
                          "Country":"City","Working Experience In This Firm":"Experience"})
df1.columns
```

```python
df2 = df1.copy()
```

```python
df2 = df2.dropna(subset=['Department'])
print(df2.isna().sum())
print('*' * 100)
print("Shape: ",df2.shape)
```

```python
print('*' * 100)
print("Unique: ",df2['Age'].unique())
print('*' * 100)
print("Mean: ",df2['Age'].mean())
print('*' * 100)
print("Median: ",df2['Age'].median())
print('*' * 100)
print("Mode: ",df2['Age'].mode())
print('*' * 100)
print("Description: \n",df2['Age'].describe())
print('*' * 100)
```

```python
df2['Age'].fillna(df2['Age'].median(), inplace = True)
print(df2.isnull().sum())
print('*' * 100)
print("Shape: ",df2.shape)
```

```python
print('*' * 100)
print("Unique: ",df2['Experience'].unique())
print('*' * 100)
print("Mean: ",df2['Experience'].mean())
print('*' * 100)
print("Median: ",df2['Experience'].median())
print('*' * 100)
print("Mode: ",df2['Experience'].mode())
print('*' * 100)
print("Description: \n",df2['Experience'].describe())
print('*' * 100)
```

```python
df2['Experience'].fillna(df2['Experience'].median(), inplace = True)
print(df2.isnull().sum())
print('*' * 100)
print("Shape: ",df2.shape)
```

```python
Experience_categories = df2['Experience']
Experience_categories = Experience_categories.dropna()

bins = [0,4,13,18,25,Experience_categories.max()]
binlabels = ['Entry Level', 'Associate', 'Manager', 'Director', 'ShareHolder']

categories = pd.cut(Experience_categories,bins,labels = binlabels)
categories.tail(20)
```

```python
df2['ExperienceCategories'] = categories
df2.tail(15)
```

```python
df3 = df2.copy()
```

```python
Agegreaterthan35andDepartmentasIT = df3[(df3['Age'] > 25) & (df3['Department'] == 'IT')]
print("Shape: ",Agegreaterthan35andDepartmentasIT.shape)
Agegreaterthan35andDepartmentasIT
```

```python
Agegreaterthan35andExpCategoryasManager = df3[(df3['Age'] > 25) & (df3['ExperienceCategories'] == 'Manager')]
print("Shape: ",Agegreaterthan35andExpCategoryasManager.shape)
Agegreaterthan35andExpCategoryasManager
```

```python
CityinandExpCategoryasDirector = df3[((df3['City'] == 'New york') | (df3['City'] == 'San Francisco')) & (df3['ExperienceCategories'] == 'Director')]
print("Shape: ",CityinandExpCategoryasDirector.shape)
CityinandExpCategoryasDirector
```
