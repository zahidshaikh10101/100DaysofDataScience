# NLTK

```python
import nltk

# Sample text
text = """
Warren Edward Buffett was born on August 30, 1930. He is an American businessman, investor, and philanthropist who currently serves as the chairman and CEO of Berkshire Hathaway. As a result of his investment success, Buffett is one of the best-known investors in the world. As of October 2024, he had a net worth of $147 billion, making him the eighth-richest person in the world
Buffett was born in Omaha, Nebraska. The son of US congressman and businessman Howard Buffett, he developed an interest in business and investing during his youth. He entered the Wharton School of the University of Pennsylvania in 1947 before graduating from the University of Nebraska at 19. He went on to graduate from Columbia Business School, where he molded his investment philosophy around the concept of value investing pioneered by Benjamin Graham. He attended New York Institute of Finance to focus on his economics background and soon pursued a business career.
"""

# Tokenize and POS tagging
tokens = nltk.word_tokenize(text)
tags = nltk.pos_tag(tokens)
print(tokens)
print('-' * 150)
print('-' * 150)
print(tags)
print('-' * 150)
print('-' * 150)
# Named entity recognition
named_entities = nltk.ne_chunk(tags)
print(named_entities)
print('-' * 150)
print('-' * 150)

# Separate entities by label and remove duplicates
person_list = set()
organization_list = set()
gpe_list = set()

for entity in named_entities:
    if hasattr(entity, "label"):
        entity_text = " ".join(c[0] for c in entity)
        entity_label = entity.label()
        if entity_label == "PERSON":
            person_list.add(entity_text)
        elif entity_label == "ORGANIZATION":
            organization_list.add(entity_text)
        elif entity_label == "GPE":
            gpe_list.add(entity_text)

# Convert sets to lists for further use
person_list = list(person_list)
organization_list = list(organization_list)
gpe_list = list(gpe_list)

# Print results
print("Unique PERSON Entities:", person_list)
print("Unique ORGANIZATION Entities:", organization_list)
print("Unique GPE Entities:", gpe_list)
```
